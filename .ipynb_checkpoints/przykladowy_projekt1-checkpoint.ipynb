# Rozpoznawanie twarzy i pisma odręcznego

## Eigenfaces (twarze własne)

Problem postawiony jest następująco: mamy zdjęcia kilku osób (np. w lepszej i gorszej jakości, w różnym wieku, z różnynmi minami ) i chcemy na ich podstawie stworzyć narzędzie do identyfikacji osób na podstawie innych zdjęć. Jest to typowy problem klasyfikacyjny - każda z osób ze zdjęć jest osobną klasą, a zdjęcie, które będziemy chcieli identyfikować - elementem zbioru testowego, który ma zostać przypisany do jednej z klas z określonym błędem.

Jak sama nazwa wskazuje - algorytm wykorzystuje wektory własne. Zdjęcia traktowane są jako macierze pikseli, z których każdy piksel ma określony kolor, chociaż bardzo często, dla ułatwienia klasyfikacji, zdjęcia poddawane są najpierw desaturacji (konwertowane na odcienie szarości), a więc macierz reprezentująca zdjęcie jest najzwyczajniej mniejsza. Brzmi to zupełnie jak opis danych, które wykorzystywaliśmy przy PCA. PCA jest tutaj wstępnym etapem obróbki danych - wykorzystywane jest do redukcji wymiarów. Twarze, jakie generuje PCA zwane są właśnie twarzami własnymi i są one konstruowane dla zbioru treningowego. W tej części zajęć zobaczymy jak wyglądać będzie twarz własna zbudowana na podstawie zbioru uczącego.

Algorytm ten może zostać wykorzystany nie tylko do rozpoznawania twarzy, ale również pisma czy innych zdjęć (np. obiektów na zdjęciu, takich jak budynki, znaki drogowe, rowery, zwierzęta itp.).

Prześledźmy, jak działa rozpoznawanie twarzy metodą Eigenfaces na przykładowym zbiorze zdjęć. Tym razem nie skorzystamy z gotowej implementacji PCA, a wykorzystamy własną. Celem wyznaczenia wektorów własnych i wartości własnych skorzystamy z algorytmu rozkładu SVD zaimplementowanego w bibliotece numpy.

Zdjęcia zapisane będą w formacie PGM - prostej, tekstowej reprezentacji obrazów. Osoby zainteresowane tym formatem odsyłam pod adres  https://pl.wikipedia.org/wiki/Portable_anymap, gdzie znaleźć można krótkie, lecz jasne wytłumaczenie idei formatu PGM.

```python
import matplotlib.pyplot as plt
import numpy as np
import os

def plot_portraits(images, titles, h, w, n_row, n_col):
    plt.figure(figsize=(2.2 * n_col, 2.2 * n_row))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.20)
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i])
        plt.xticks(())
        plt.yticks(())
        
def pca(X, n_pc):
    n_samples, n_features = X.shape
    mean = np.mean(X, axis=0)
    centered_data = X-mean
    U, S, V = np.linalg.svd(centered_data)
    components = V[:n_pc]
    projected = U[:,:n_pc]*S[:n_pc]
    
    return projected, components, mean, centered_data

def reconstruction(Y, C, M, h, w, image_index):
    n_samples, n_features = Y.shape
    weights = np.dot(Y, C.T)
    centered_vector=np.dot(weights[image_index, :], C)
    recovered_image=(M+centered_vector).reshape(h, w)
    return recovered_image
```

Najpierw wyświetlmy kilka z obrazów w oryginale (cały zestaw plików pgm to ponad 13 000 obrazów - są to twarze osób znanych, wycięte z różnych zdjęć - dla każdej z osób jest kilka zdjęć). Dodatkowo - wykorzystamy tylko 1000 zdjęć - można próbować później zabawy z większą liczbą zdjęć - zwiększy to oczywiście czas wykonywanych obliczeń i złożoność pamięciową. Z 1000 fotografii nie powinien mieć problemów zwykły, domowy PC.

```python
dir='./Dane/lfwcrop_grey/faces'
celebrity_photos=os.listdir(dir)[1:1001]
celebrity_images=[dir+'/' + photo for photo in celebrity_photos]
images=np.array([plt.imread(image) for image in celebrity_images], dtype=np.float64)
celebrity_names=[name[:name.find('0')-1].replace("_", " ") for name in celebrity_photos]
n_samples, h, w = images.shape
plot_portraits(images, celebrity_names, h, w, n_row=4, n_col=4)
```

Następnie wykonajmy redukcję wymiarów do $n=50$ - polecam Państwu przetestować ten krok ze zmianą liczby wymiarów, aby zobaczyć jak wyglądają twarze własne z mniejszą i większą liczbą składowych. W wyniku działania poniżego kodu otrzymamy 50 twarzy własnych.

```python
n_components = 50
X = images.reshape(n_samples, h*w)
P, C, M, Y= pca(X, n_pc=n_components)
eigenfaces = C.reshape((n_components, h, w))
eigenface_titles = ["eigenface %d" % i for i in range(eigenfaces.shape[0])]
plot_portraits(eigenfaces, eigenface_titles, h, w, 4, 4) 
```

### Rekonstrukcja twarzy na podstawie twarzy własnych

Teraz czas na rekonstrukcję twarzy na podstawie twarzy własnych. Jest to nic innego jak rzutowanie oryginalnych obrazów na nowe wektory bazowe, które wyznaczone zostały przez PCA.

```python
recovered_images=[reconstruction(Y, C, M, h, w, i) for i in range(len(images))]
plot_portraits(recovered_images, celebrity_names, h, w, n_row=4, n_col=4)
```

Każda z powyższych twarzy jest ważoną kombinacją twarzy własnych. Polecam porównać te twarze z oryginałami. Warto zwrócić uwagę na fakt, że wynikowe twarze mają zdecydowanie mniejszy rozmiar - z 1000 pikseli zeszliśmy do 50 pikseli - jest to 20 razy mniej danych niż w oryginale! Na tym etapie - polecam również zobaczyć jaki efekt rekonstrukcji otrzymamy przy mniejszej lub większej liczbie składowych w algorytmie PCA. 

### Rozpoznawanie twarzy

Teraz skupimy się na rozpoznawaniu twarzy, a więc klasyfikacji. Pomimo, że w poprzednim przykładzie skorzystaliśmy z własnej implementacji PCA, tym razem dla ułatwienia skorzystamy z gotowej implementacji w pakiecie sklearn - dokładnie tej samej, z której korzystaliśmy przy redukcji wymiarów na poprzednich zajęciach.

Pełen zbiór twarzy dostarczy nam gotowy zbiór danych z pakietu sklearn (UWAGA! Zbiór jest duży i nie znajduje się w repozytorium - pierwsze uruchomenie kodu wymaga ściągnięcia danych).

Zbiór podzielimy na dwie części - dane treningowe i testowe w losowy sposób. Załóżmy, że 70% danych to zbiór treningowy, a 30% to zbiór testowy.

Jako klasyfikatora użyjemy trochę bardziej skomplikowanej metody, jaką jest MLP (ang. Multi Layer Perceptron), a więc wielowymiarowy perceptron. Jego działanie i szczegóły implementacji sa poruszone w ramach kursu ze Machine learning Sztucznych Sieci Neuronowych - dzisiaj, będzie to dla nas "czarna skrzynka", która dokona klasyfikacji.

```python
import matplotlib.pyplot as plt
 
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_lfw_people
from sklearn.metrics import classification_report
from sklearn.decomposition import PCA
from sklearn.neural_network import MLPClassifier
 

lfw_dataset = fetch_lfw_people(min_faces_per_person=100)
 
_, h, w = lfw_dataset.images.shape
X = lfw_dataset.data
y = lfw_dataset.target
target_names = lfw_dataset.target_names
 
# split into a training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
n_components = 100
pca = PCA(n_components=n_components, whiten=True).fit(X_train)
 
# apply PCA transformation
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

# train a neural network
print("Dopasowanie klasyfikatora do danych ze zbioru testowego")
clf = MLPClassifier(hidden_layer_sizes=(1024,), batch_size=256, verbose=True, early_stopping=True).fit(X_train_pca, y_train)
```

```python
y_pred = clf.predict(X_test_pca)
print(classification_report(y_test, y_pred, target_names=target_names))
```

Powyższe podsumowanie pokazuje efekty klasyfikacji - widać, że tam gdzie było dużo danych dla klasy (kolumna support) - wyniki klasyfikacji są bardzo dobre. Precision to oczywiście precyzja klasyfikacji, a recall - stosunek poprawnych klasyfikacji do wszytskich. F1-score wyznaczany jest na podstawie dwóch poprzednich parametrów.

Pozostaje tylko wyświetlić wyniki w bardziej przyjemnej, graficznej formie.

```python
def plot_gallery(images, titles, h, w, rows=3, cols=4):
    plt.figure()
    for i in range(rows * cols):
        plt.subplot(rows, cols, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i])
        plt.xticks(())
        plt.yticks(())
        plt.subplots_adjust(hspace=1, wspace=1)
 
def titles(y_pred, y_test, target_names):
    for i in range(y_pred.shape[0]):
        pred_name = target_names[y_pred[i]].split(' ')[-1]
        true_name = target_names[y_test[i]].split(' ')[-1]
        yield 'P: {0}\nT: {1}'.format(pred_name, true_name)
 
prediction_titles = list(titles(y_pred, y_test, target_names))
plot_gallery(X_test, prediction_titles, h, w)
```

## Fisherfaces (twarze Fishera)

Podobną metodą do twarzy własnych są twarze Fishera. Opierają się one na algorytmie LDA, zwanym również algorytmem Fishera. Jest to nawet starsza metoda, która użyta została do redukcji wymiarowości i klasyfikacji zbiorów IRIS.

Twarze Fishera różnią się od twarzy własnych tym, że metoda LDA maksymalizuje stosunek odległości pomiędzy klasami do rozproszenia wewnątrz klas.

Wygląd twarzy Fishera utworzonej ze zbioru uczącego niekoniecznie przypomina "uśrednioną twarz", tak jak miło to miejsce w omawianej wcześniej metodzie. Wynika to właśnie z faktu, że algorytm identyfikuje obszary twarzy, które są od siebie jak najbardziej oddalone.
