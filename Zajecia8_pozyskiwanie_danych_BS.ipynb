```python
#-*- coding: utf-8 -*-

import urllib.request
from bs4 import BeautifulSoup
import re
import pandas as pd


URLS = ['https://www.tekstowo.pl/piosenki_artysty,bracia_figo_fagot.html', 
'https://www.tekstowo.pl/piosenki_artysty,bracia_figo_fagot,alfabetycznie,strona,2.html',
'https://www.tekstowo.pl/piosenki_artysty,bracia_figo_fagot,alfabetycznie,strona,3.html']

def dane(URL):
  page = urllib.request.urlopen(URL)
  soup = BeautifulSoup(page)
  temp = soup.body.findAll('div', attrs={'class' : 'box-przeboje'}) # zeby dowiedziec do jakiego elementu musimy sie odniesc, trzeba zajrzec do wnetrza strony
  return temp

dane(URLS[0]) #dostajemy wszystkie divy z class box-przeboje
```

```python

from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import urllib.request as urlrq
import certifi






url = "https://www.imdb.com/title/tt0468569/fullcredits"
#page = urlopen(url)
page = urlrq.urlopen(url, cafile=certifi.where())

html = page.read().decode("utf-8")
zupa = BeautifulSoup(html, "html.parser")



mytds = zupa.findAll("td", {"class": "primary_photo"})


mytds[1].select('img')[0].attrs['title']

list_actors = [mytds[i].select('img')[0].attrs['title'] for i in range(len(mytds))]
list_href = ['https://www.imdb.com/'+mytds[i].select('a')[0]['href'] for i in range(len(mytds))]





def get_actors_by_movie_soup(cast_page_soup, num_of_actors_limit=None):
  soup =cast_page_soup
  mytds = soup.findAll("td", {"class": "primary_photo"})
  list_actors = [mytds[i].select('img')[0].attrs['title'] for i in range(len(mytds))]
  list_href = ['https://www.imdb.com/'+mytds[i].select('a')[0]['href'] for i in range(len(mytds))]


  # tag_urls = soup.select('a[data-testid="title-cast-item__actor"]')
  # list_actors = [x.text for x in tag_urls]
  # list_href = ['https://www.imdb.com/'+x['href'] for x in tag_urls]


  pairs_actors = list(zip(list_actors, list_href))

  if num_of_actors_limit==None:
    return pairs_actors
  else:
    return pairs_actors[:num_of_actors_limit]


# url = "https://www.imdb.com/name/nm0000288/?ref_=tt_cl_t_1"
url = "https://www.imdb.com/name/nm0367176/?ref_=tt_cl_t_18"
    
# page = urlopen(url)
page = urlrq.urlopen(url, cafile=certifi.where())
html = page.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")



def get_movies_by_actor_soup(actor_page_soup, num_of_movies_limit=None):
    
    zupa=actor_page_soup.findAll(True, attrs={"class":"filmo-category-section"})[0].findAll(True, {"class":["filmo-row even", "filmo-row odd"]})
    link_list = []
    name_list = []
    for i in range(len(zupa)):
      prod = zupa[i].select("a[class]")


      if len(prod)>0 and prod[0]['class'][0] == 'in_production' or len(zupa[i].find('b').next_sibling.strip())>1:
        continue
      else:

        movie_name = zupa[i].select("b")[0].text

        short = list(zupa[i].select('a[href]')[0].attrs.values())[0]
        link = "https://www.imdb.com/{short}".format(short=short)
        link_list.append(link)
        name_list.append(movie_name)
    pairs_movies = list(zip(link_list, name_list))

    if num_of_movies_limit==None:
      return pairs_movies
    else:
      return pairs_movies[:num_of_movies_limit]



print(get_movies_by_actor_soup(soup, 5))
#get_movies_by_actor_soup(soup)
#movie_cast = get_actors_by_movie_soup(zupa)
#print(*movie_cast[:20], sep='\n')



#get_actors_by_movie_soup(zupa)


```
